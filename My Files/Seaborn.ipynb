```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'],kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data= tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,palette='pink',hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True,hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='sex',y='total_bill',data=tips)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
crr = tips.corr()
```

```python
sns.heatmap(crr,cmap='coolwarm',annot=True)
```

```python
pf = flights.pivot_table(values='passengers',index ='month',columns='year')
```

```python
sns.heatmap(pf)
```

```python
sns.clustermap(pf,cmap='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', col='sex', row='time')
```

```python
iris  = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips,col='time',row='smoker')
g.map(plt.hist,'total_bill')
```
