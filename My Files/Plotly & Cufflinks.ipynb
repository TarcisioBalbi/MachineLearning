```python
data = dict(type='choropleth',
            locations =['AZ','CA','NY'],
            locationmode='USA-states',
            colorscale='Portland',
            text=['Texto1','Texto2','Texto3'],
            z=[1.0,2.0,3.0],
            marker=dict(line=dict(color='rgb(255,255,255)',width=2)),
            colorbar={'title':'Titulo da barra de cores'}           )
```

```python
layout = dict(geo={'scope':'usa'})
```

```python
choromap = go.Figure(data=[data],layout=layout)
iplot(choromap)
```

```python
df = pd.read_csv('2011_US_AGRI_Exports')
df.head()
```

```python
data = dict(type='choropleth',
            locations =df['code'],
            locationmode='USA-states',
            colorscale='ylorrd',
            text=df['text'],
            z=df['total exports'],
            marker=dict(line=dict(color='rgb(0,0,0)',width=1)),
            colorbar={'title':'Milhoes de dolares'})
```

```python
layout =dict(title='Exportações agrícolas de 2011 por estado',geo=dict(scope='usa',showlakes=True,lakecolor='rgb(85,173,240)'))
```

```python
choromap = go.Figure(data=[data],layout=layout)
iplot(choromap)
```

```python
df = pd.read_csv('2014_world_GDP')
df.head()
```

```python
data = dict(type='choropleth',
            locations =df['CODE'],
            
            colorscale='ylorrd',
            text=df['COUNTRY'],
            z=df['GDP (BILLIONS)'],
            marker=dict(line=dict(color='rgb(0,0,0)',width=1)),
            colorbar={'title':'PIB em Milhões'})
```

```python
layout=dict(title='PIB mundo 2014', geo=dict(showframe=False,projection ={'type':'mercator'}))
```

```python
choromap=go.Figure(data=[data],layout=layout)
iplot(choromap)
```
